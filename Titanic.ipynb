```

#test = 1tvhsydxmYuaQgfDsiMqewMvUxC_0fx0L
#train = 1GBYiHKwmjhe_-487NAGfMsYTJo3bYIGG

train_downloaded = drive.CreateFile({'id':'1GBYiHKwmjhe_-487NAGfMsYTJo3bYIGG'})
train_downloaded.GetContentFile('train.csv')
test_downloaded = drive.CreateFile({'id':'1tvhsydxmYuaQgfDsiMqewMvUxC_0fx0L'})
test_downloaded.GetContentFile('test.csv')

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

#df_train.head()
df_train.describe()
#df_test.head()
```

```
#Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
y = df_train.pop("Survived")
y.head()
```

```
numeric_variables = list(df_train.dtypes[df_train.dtypes != "object"].index)
df_train[numeric_variables].head() 
```

```
df_train["Age"].fillna(df_train.Age.mean(), inplace=True)
df_train.tail()
```

```
df_train[numeric_variables].head()
```

```
model = RandomForestClassifier(n_estimators=100)
model.fit(df_train[numeric_variables], y)
```

```
a = model.predict(df_train[numeric_variables])
#print (a)

print ("Train Accuracy :: ", accuracy_score(y,a))
```

```
df_test[numeric_variables].head()
```

```
df_test["Age"].fillna(df_test.Age.mean(), inplace=True)
```

```
df_test = df_test[numeric_variables].fillna(df_test.mean()).copy()
```

```
y_pred = model.predict(df_test[numeric_variables])
y_pred
```

```
submission = pd.DataFrame({
    "PassengerId": df_test["PassengerId"],
    "Survived": y_pred
})

submission.to_csv('titanicPresentation.csv', index=False, sep='\t')
submission.head()
```
